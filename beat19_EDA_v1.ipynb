```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes = True)

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)

from uszipcode import SearchEngine
```

```python
daily = pd.read_csv("beat19-collab-data\\versions\\2020-06-01\\daily-data.csv")
enrollment = pd.read_csv("beat19-collab-data\\versions\\2020-06-01\\enrollment-data.csv")
```

```python
daily.info()
```

```python
total_records = len(daily)
unique_ids = daily._id.nunique()
print(f"number of records: {total_records}" )
print(f"number of unique _ids': {unique_ids}")
```

**Average number of entries per participant:**

```python
avg_time_steps = total_records / unique_ids
print(f"each unique id has an average of{avg_time_steps: .0f} records")
```

**NaN percent per feature:**

```python
percent_nan = daily.isna().sum() * 100 / len(daily)
missing_values = pd.DataFrame({'column_name' : daily.columns,
                              'percent_nan' : percent_nan})
missing_values
```

```python
daily[daily['version'] == 1]
```

```python
percent_nan_1 = daily[daily['version'] == 1].isna().sum() * 100 / len(daily[daily['version'] == 1])
percent_nan_2 = daily[daily['version'] == 2].isna().sum() * 100 / len(daily[daily['version'] == 2])
missing_values = pd.DataFrame({'column_name' : daily.columns,
                              'percent_nan_1' : percent_nan_1,
                              'percent_nan_2' : percent_nan_2})
missing_values.reset_index()
```

```python
percent_of_total_1 = daily[daily['version'] == 1].notna().sum() * 100 / daily.notna().sum()
percent_of_total_2 = daily[daily['version'] == 2].notna().sum() * 100 / daily.notna().sum()
percent_of_values = pd.DataFrame({'column_name' : daily.columns,
                              'percent_of_total_1' : percent_of_total_1,
                              'percent_of_total_2' : percent_of_total_2})
percent_of_values.reset_index()
```
